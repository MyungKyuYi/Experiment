```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
import matplotlib.pyplot as plt

# 1. 데이터 로드
file_path = 'E:/dataset/Hypertension/Kaggle/D1_P_26_2000.csv'
df = pd.read_csv(file_path)

# 2. 특징과 레이블 분리
feature_cols = [f'PPG_{i}' for i in range(2000)]
X = df[feature_cols].values
y = df['BP_class'].values

# 3. 차원 축소 기법별 변환
# 3-1. PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 3-2. t-SNE
tsne = TSNE(n_components=2, random_state=42, init='pca')
X_tsne = tsne.fit_transform(X)

# 3-3. UMAP
umap_reducer = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_reducer.fit_transform(X)

# 4. 결과 시각화 (각각 크게)
# PCA
plt.figure(figsize=(10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, s=10, alpha=0.7)
plt.title('PCA', fontsize=16)
plt.xlabel('PC1', fontsize=14)
plt.ylabel('PC2', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

from sklearn.manifold import TSNE

tsne = TSNE(
    n_components=2,
    random_state=42,
    init='random',      # PCA 대신 완전 랜덤 초기화
    learning_rate='auto',
    perplexity=30
)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(10,8))
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y, s=10, alpha=0.7)
plt.title('t-SNE', fontsize=16)
plt.xlabel('Dim 1', fontsize=14)
plt.ylabel('Dim 2', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

# UMAP
plt.figure(figsize=(10, 8))
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y, s=10, alpha=0.7)
plt.title('UMAP', fontsize=16)
plt.xlabel('UMAP1', fontsize=14)
plt.ylabel('UMAP2', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

```
